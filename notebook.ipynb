```python
import pandas as pd
import numpy as np
import json
import ast
```

```python
credits = pd.read_csv(r"Data\tmdb_5000_credits.csv")
movies = pd.read_csv(r"Data\tmdb_5000_movies.csv")
```

```python
credits.head()
```

```python
movies.head()
```

```python
movies.info()
```

```python
# For merging on different columns
# pd.merge(movies, credits, how='left', left_on='id', right_on='movie_id')
Movies = pd.merge(movies, credits, on='title')
Movies.head()
```

```python
Movies = Movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
Movies.head()
```

```python
Movies.isnull().sum()
```

```python
Movies.dropna(inplace=True)
```

```python
Movies.duplicated().sum()
```

```python
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
Movies['genres'] = Movies['genres'].apply(convert)
Movies.head()
```

```python
Movies['keywords'] = Movies['keywords'].apply(convert)
Movies.head()
```

```python
def convert_1(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        else:
            break
        counter +=1
    return(L)
    
```

```python
Movies['cast'] = Movies['cast'].apply(convert_1)
Movies.head()
```

```python
def find_director(text):
    l= []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
Movies['crew'] = Movies['crew'].apply(find_director)
Movies.head()
```

```python
Movies['overview']= Movies['overview'].apply(lambda x : x.split(' '))
```

```python
Movies['genres'] = Movies['genres'].apply(lambda x : [i.replace(" ","") for i in x])
Movies['keywords'] = Movies['keywords'].apply(lambda x : [i.replace(" ","") for i in x])
Movies['cast'] = Movies['cast'].apply(lambda x : [i.replace(" ","") for i in x])
Movies['crew'] = Movies['crew'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
Movies['tag'] = Movies['overview'] + Movies['genres'] + Movies['keywords'] + Movies['cast'] + Movies['crew']
Movies.head()
```

```python
new = Movies.drop(columns=['overview', 'genres', 'keywords', 'cast', 'crew'])
new.head()
```

```python
new['tag'] = new['tag'].apply(lambda x : " ".join(x))
new['tag'] = new['tag'].apply(lambda x : x.lower())
new.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
ps.stem('loving')
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new['tag'] = new['tag'].apply(stem)
new.head()
```

```python
from sklearn. feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(new['tag']).toarray()
```

```python
import sys
np.set_printoptions(threshold=sys.maxsize) # to get non-truncated output of array
```

```python
vector.shape
```

```python
cv.get_feature_names_out()
```

```python
# Not to calculate eucradin distance, but cosine angle. For high dimension, use cosine angle.
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
a = sorted(list(enumerate(similarity[0])), reverse=True, key= lambda x : x[1])[1:6]
[new.iloc[i[0]]['title'] for i in a]
```

```python
new[new["title"] == "El Mariachi"].index[0]
```

```python
def recommend(movies_title):
    movies_index = new[new["title"] == movies_title].index[0]
    distances = similarity[movies_index]
    closet = sorted(list(enumerate(distances)), reverse=True, key= lambda x : x[1])
    [print(new.iloc[i[0]]['title']) for i in closet[1:6]]
```

```python
recommend('Batman Begins')
```
